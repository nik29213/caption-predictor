```python
def preprocess_img(img):
    img = image.load_img(img,target_size=(224,224))
    img = image.img_to_array(img)
    img = np.expand_dims(img,axis=0) 
    #bcz img may b needed to b feed in batches .so its dim will become(n,224,224,3).so expand_im is used
    img = preprocess_input(img)#predefined func of keras
    return img
```

```python
def encode_img(img):
    img = preprocess_img(img)
    feature_vector = new_model.predict(img)
    feature_vector = feature_vector.reshape((-1,))
    return feature_vector
```

### caption preprocessing

```python
word2idx = {}
idx2word = {}
for idx,word in enumerate(total_words):
    word2idx[word] = idx+1
    idx2word[idx+1] = word
print(len(idx2word))   
```

```python
word2idx["startseq"] = 1851
idx2word[1851] = "startseq"

word2idx["endseq"] = 1852
idx2word[1852] = "endseq"
```

```python
vocab_size = len(idx2word)+1
vocab_size
```

```python
max_len = 0
for key in train_description.keys():
    for cap in train_description[key]:
        max_len = max(max_len,len(cap.split()))
print(max_len)
```

## data loader- Generator

```python
def data_generator(train_description,encode_train,word2idx,max_len,batch_size):
    x1,x2,y = [],[],[]
    n=0
    while True:
        for key,desc_list in train_description.items():
            n+=1
            photo = encode_train[key]
            
            for desc in desc_list:
                seq = [word2idx[word] for word in desc.split() if word in word2idx]
                for i in range(1,len(seq)):
                    xi = seq[0:i]
                    yi = seq[i]
                    
                    xi = pad_sequences([xi],maxlen = max_len,value = 0,padding = "post")[0]
                    #0 will be added as padding value after the nos. in seq if its less than max_len
                    
                    yi = to_categorical([yi],num_classes = vocab_size)[0]
                    
                    x1.append(photo)
                    x2.append(xi)
                    y.append(yi)
                    
                if(n == batch_size):
                    yield [[np.array(x1),np.array(x2)],np.array(y)]
                    
                    x1,x2,y = [],[],[]
                    n=0
        
```

## word embedding

```python
f = open("../emoji_predictor/glove.6B.50d.txt",encoding = "utf-8")
```

```python
embedding_index = {}
for line in f:
    values = line.split()
    word = values[0]
    embedding_word = np.array(values[1:],dtype='float32')
    embedding_index[word] = embedding_word
    
```

```python
f.close()
```

```python
embedding_index["apple"]
```

```python
def get_embedding_matrix():
    emb_dim = 50
    matrix = np.zeros((vocab_size,emb_dim))
    for word,idx in word2idx.items():
        embedding_vector = embedding_index.get(word)
        
        if embedding_vector is not None:
            matrix[idx] = embedding_vector
            
    return matrix
```

```python
embedding_matrix = get_embedding_matrix()
print(embedding_matrix.shape)
```

## Model architecture

```python
# for images -extract features and convert to dim 256
img_features = Input(shape = (2048,))
input_img1 = Dropout(0.3)(img_features)
input_img2 = Dense(256,activation = "relu")(input_img1)
```

```python
# for caption - extract features and convert to dim 256
cap_features = Input(shape = (max_len,))
input_cap1 = Embedding(input_dim = vocab_size,output_dim = 50,mask_zero = True)(cap_features)
input_cap2 = Dropout(0.3)(input_cap1)
input_cap3 = LSTM(256)(input_cap2)
```

```python
decoder1 = add([input_img2,input_cap3])
decoder2 = Dense(256,activation = "relu")(decoder1)
outputs_prob = Dense(vocab_size,activation = "sigmoid")(decoder2)

#final model
model = Model(inputs = [img_features,cap_features],outputs = outputs_prob)
```

```python
model.summary()
```

```python
# pre initialize embedding layer of the above model i.e 3rd layer i.e index no.=2
model.layers[2].set_weights([embedding_matrix])
model.layers[2].trainable = False
```

```python
model.compile(loss = "categorical_crossentropy",optimizer = "adam")
```

### training

```python
epochs = 10
batch_size = 3
steps = len(train_description)//batch_size
```

```python
def train():
    for i in range(epochs):
        generator = data_generator(train_description,encode_train,word2idx,max_len,batch_size)
        model.fit_generator(generator,epochs = 1,steps_per_epoch = steps,verbose = True)
        model.save("model_weights/model_"+str(i)+".h5")
```

```python
train()
```

## predict result

```python
def predict_caption(photo):
    in_text = "startseq"
    for i in range(max_len):
        sequence = [word2idx[w] for w in in_text.split() if w in word2idx]
        sequence = pad_sequences([sequence],maxlen = max_len,padding = "post")
        #print(sequence)
        ypred = model.predict([photo,sequence])
        #print("ypred = ",str(ypred))

        ypred = ypred.argmax()
        #print("ypred = ",str(ypred))
        word = idx2word[ypred]
        in_text+=(" " + word)
        if word == "endseq":
            break
    final_caption = in_text.split()[1:-1]#remove startseq nd endseq
    final_caption = " ".join(final_caption)
        
    return final_caption
```
